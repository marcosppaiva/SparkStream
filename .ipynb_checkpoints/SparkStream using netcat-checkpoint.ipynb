# Spark Streaming

## Simulating TCP / IP data flow

```python
from pyspark.streaming import StreamingContext
```

```python
# Creating a StreamContex with a 1 second interval
ssc = StreamingContext(sc, 1)
```

```python
# Creating a DStream for connect to port 9797 in your local machine
lines = ssc.socketTextStream("localhost", 9797)
```

```python
# Divide line into words
words = lines.flatMap(lambda line : line.split(" "))
```

```python
# Printing the words
words.pprint()
```

```python
# Starts data collection and stream processing
ssc.start()
ssc.awaitTermination()
```

```python
# Close connection
ssc.stop()
```
